# <span style="color:purple">Observing Mechanistic Model Results</span>

### Load packages and data

```python
import altair as alt
import pandas as pd

# Mechanistic Model Results
mm_results = pd.read_csv('../sample_datasets/mol_res_scan_results_10_sourceQ.csv')
mm_results = mm_results.sample(5000)
mm_results = mm_results.reset_index()
```

### Observe purity vs. yield

```python
alt.Chart(mm_results).mark_point().encode(
    x='yield',
    y='purity')
```

### Optimized conditions when purity and yield approach 1.0

### Set threshhold for each purity and yield

```python
purity_thresh = 0.9
yield_thresh = 0.9
```

# <span style="color:purple">List results that exceed the threshholds set</span>
Here you can list the results from the mechanistic model that surpass the threshholds set for yield and purity

```python
# Initialize results dataframe
column_names = list(mm_results.columns.values)
optimized_runs = pd.DataFrame(columns=column_names)

# Add optimized runs to optimized array from results array
for i in range(5000):
    if mm_results['yield'].iloc[i] >= yield_thresh:
        if mm_results['purity'].iloc[i] >= purity_thresh:
            optimized_runs = optimized_runs.append(mm_results.loc[i])
            
# View optimized runs results as Pandas DataFrame 
optimized_runs
```
