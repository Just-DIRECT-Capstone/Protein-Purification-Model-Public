# Observing MM Results

### Load packages and data

```python
import altair as alt
import numpy as np
import pandas as pd

# Mechanistic Model Results
mm_results = pd.read_csv('mol_res_scan_results_7_sourceQ.csv')
mm_results = mm_results.sample(5000)
```

### Observe purity vs. yield

```python
alt.Chart(mm_results).mark_point().encode(
    x='yield',
    y='purity')
```

### Optimized conditions when purity and yield approach 1.0

### Set threshhold for each purity and yield

```python
purity_thresh = 0.9
yield_thresh = 0.9
```

### List results that exceed the threshholds set

```python
# Make dataframe into numpy array
mm_array = mm_results.to_numpy()

# Initialize results array
optimized_array = np.empty((1,21))

# Add optimized runs to optimized array from results array
for i in range(5000):
    if mm_array['yield'][i] >= yield_thresh:
        if mm_array['purity'][i] >= purity_thresh:
            optimized_array.append(mm_array[i])
        else:
            continue
    else:
        continue

# Convert results array to dataframe
optimized_runs = pd.DataFrame(optimized_array)
```
